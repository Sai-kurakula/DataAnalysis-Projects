#Individuals Using Internet over years in the World. (1990-2020)



```python
import pandas as pd
mydata = pd.read_csv('/Users/saikurakula/Downloads/internetusage.csv')
mydata.head()
```

```python
#subset the data of a specific year:

usage2019 = mydata.query("Year == 2019")

usage2019.head()
```

```python
import plotly.express as px
```

```python
fig= px.choropleth(
                   usage2019,
                   locations = 'Code',
                   color = "Individuals using the Internet (% of population)",
                   hover_name = "Entity",
                   color_continuous_scale = px.colors.sequential.Plasma
                   )
fig.show()
```

```python
fig.update_layout(
    title_text = "Internet usage across the world(% Population) - 2016"
)

fig.show()
```

```python
fig.update_layout(
    title_text = "Internet usage across the world (% Population) - 2016",
    geo = dict(projection = {'type': 'natural earth'})) # see the projection, by default its set to "equirectangular"


fig.show()
```

```python
fig.update_layout(
    title_text = "internet usage across the world (% Population) - 2016",
    geo = dict(projection_type = "orthographic"))

fig.show()
```

```python
mydata.sort_values(by = ['Year'], inplace = True)
#if inplace is set to True, it performs the operation in place, modifies the existing dataframe & you dont need to assign a new data frame.

mydata.head()
```

```python
import plotly.express as pandas

fig = px.choropleth(mydata, locations = 'Code',
                     color = "Individuals using the Internet (% of population)",
                     color_continuous_scale = px.colors.sequential.Plasma,
                     animation_frame = 'Year')
fig.update_layout(
    title_text = "Internet usage across the world (% Population) - 2016",
    geo = dict(projection_type = "natural earth")
)

fig.show()
```
